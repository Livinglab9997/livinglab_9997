```python
mask = df['상태'] == '방사'
df = df[mask]
```

```python
count_different = (df['방사장소'] != df['포획장소']).sum()
print("방사장소와 포획장소가 다른 경우의 수:", count_different)
```

```python
different = df['방사장소'] != df['포획장소']
df[different]
```

```python
different_rows = df[df['방사장소'] != df['포획장소']]

# 각 경우에 대해 상세 정보 출력
for idx in different_rows.index:
    print(f"{idx}: 방사장소 - {df['방사장소'][idx]}, 포획장소 - {df['포획장소'][idx]}")
```

```python
# '방사장소'와 '포획장소'가 다른지 여부를 나타내는 열 추가
df['방사포획일치'] = df['방사장소'] == df['포획장소']
```
