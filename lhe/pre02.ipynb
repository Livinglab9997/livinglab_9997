```python
import seaborn as sns

# 각 지역의 TNR 개체 성비 계산
tnr_ratio = df.groupby('사업지역(시군구)')['성별'].value_counts(normalize=True).unstack()


# 나비 차트 그리기
plt.figure(figsize=(10, 6))

# 수컷 그래프 그리기
plt.barh(tnr_ratio.index, tnr_ratio[0], color='skyblue', label='수컷')

# 암컷 그래프 그리기
plt.barh(tnr_ratio.index, -tnr_ratio[1], color='pink', label='암컷')

plt.xlabel('성비')
plt.title('각 지역의 TNR 개체 성비')
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 히스토그램 그리기
plt.figure(figsize=(10, 6))
sns.histplot(df['몸무게(kg)'], bins=20, kde=True)  # bins는 히스토그램의 막대 개수, kde는 커널 밀도 추정 그래프 여부
plt.xlabel('몸무게(kg)')
plt.ylabel('빈도')
plt.title('TNR 대상 개체 몸무게(kg)')
plt.grid(True)
plt.show()
```
