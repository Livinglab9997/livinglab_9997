```python
size = 10
fig_size = 15
plt.figure(figsize = (fig_size, fig_size*3/2))
for k in range(len(cols_01)):
    plt.subplot(int(len(cols_01)/2), 2, k+1)
    data[f"{cols_01[k]}"].value_counts().plot(kind='bar', colormap=plt.cm.Set2)
    plt.title(f'[ Bar chart of {cols_01[k]} ]', fontsize = size+5)
    plt.xlabel(f'{cols_01[k]}', fontsize = size)
    plt.ylabel('Frequency', fontsize = size)
    plt.xticks(rotation = 30)
    plt.grid(axis = 'y')

plt.tight_layout()

plt.show()
```
